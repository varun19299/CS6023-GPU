### Question 2

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
fig = plt.figure(figsize=(10, 10))
dpi=fig.dpi
```

```python
clock=5505000 ## kHz
clock=clock/10**6 # Ghz
peak_flops=3584*clock*2 ## TFLOPS
peak_flops=11.3
print(peak_flops)
peak_band=0.484440000 ## GB/s
x=peak_flops/peak_band
tt=np.logspace(-3,6,200,base=2)

ii=np.argmin(np.abs(tt*peak_band-peak_flops))
print(x)
print(tt*peak_band)
print(peak_flops)
```

```python
## Bandwith line
plt.plot(tt[:ii+1],peak_band*tt[:ii+1],'r');
plt.plot(tt[ii:],peak_flops*np.ones_like(tt)[ii:],'b');
plt.xlabel("Ops per Byte; Performance")
plt.ylabel("Ops per cycle; Op intensity")
plt.axis([0,60,0,12])
plt.show()
plt.savefig("q2.jpg")
```
